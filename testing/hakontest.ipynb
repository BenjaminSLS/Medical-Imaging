```python
import numpy as np
import matplotlib.pyplot as plt
from skimage.color import rgb2gray
from skimage import data
from skimage.filters import gaussian
from skimage.segmentation import active_contour
from skimage.transform import resize
```

```python
#Make iamge gray
img = plt.imread("./data/images/imgs_part_1/PAT_8_15_820.png")

im = rgb2gray(img[200:800,200:800,:3])
img = resize(im,(im.shape[0],im.shape[1]),anti_aliasing = True)
plt.imshow(img,cmap="gray")
%matplotlib inline
```

```python
#Create snake
s = np.linspace(0,2*np.pi,400)
r = 300 + 100*np.sin(s)
c = 350 + 100*np.cos(s)
init = np.array([r,c]).T
```

```python
#Smooth image
img2 = gaussian(img,3)
plt.imshow(img2,cmap="gray")
%matplotlib inline
```

```python
#Active contour segmentation
snake = active_contour(img2,init)

#Show
fig, ax = plt.subplots(figsize=(7, 7))
ax.imshow(img,cmap="gray")
ax.plot(init[:, 1], init[:, 0], '--r', lw=3)
ax.plot(snake[:, 1], snake[:, 0], '-b', lw=3)
ax.set_xticks([]), ax.set_yticks([])
ax.axis([0, img2.shape[1], img2.shape[0], 0])

plt.show()
%matplotlib inline
```
